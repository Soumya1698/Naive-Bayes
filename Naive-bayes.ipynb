```python
import pandas as pd
import numpy as np
```

# 1. Handle Data

```python
data=pd.read_csv('pima-indians-diabetes.data.csv',header=None)
```

```python
data.rename(columns={8:'label'},inplace=True)
print 'There are',data.shape[0],'records and',data.shape[1]-1,'features with 1 label'
data
```

Each record has a class value that indicates whether the patient suffered an onset of diabetes within 5 years of when the measurements were taken (1) or not (0).

# 2. spliting the data

```python
from sklearn.cross_validation import train_test_split
y = data.pop('label')
X = data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33)
```

# 3. Calculate Priors

```python
no=y_train[y_train==0].count()  #the patient did not suffer of diabetes within 5 years
yes=y_train[y_train==1].count()  #the patient did suffer of diabetes within 5 years
total_y=y_train.count()
```

calculating the probabilities of yes's and no's

```python
p_no=no/float(total_y)
p_yes=yes/float(total_y)
print p_yes,p_no
```

```python
p_no+p_yes==1
```

# 4. Calculate Likelihood

```python
X=pd.DataFrame(X_train)
X['label1']=y_train
data_means=X.groupby('label1').mean()
data_means
```

```python
data_variance=X.groupby('label1').var()
data_variance
```

# 5. Naive bayes

```python
def p_x_given_y(x, mean_y, variance_y):

    # Input the arguments into a probability density function
    p = 1/(np.sqrt(2*np.pi*variance_y)) * np.exp((-(x-mean_y)**2)/(2*variance_y))
    
    # return p
    return p
```

```python
ans=[0]*X_test.shape[0]
for j in range(X_test.shape[0]):
    ans_no=p_no
    ans_yes=p_yes
    for i in range(0,8):
        ans_no*=p_x_given_y(X_test.iloc[j][i],data_means.iloc[0][i],data_variance.iloc[0][i])
        ans_yes*=p_x_given_y(X_test.iloc[j][i],data_means.iloc[1][i],data_variance.iloc[1][i])
    ans[j]=int(ans_yes>ans_no)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ans, y_test)
```
